```python
non_top_1_perc['date']=pd.to_datetime(non_top_1_perc['date'])
```

```python
non_top_1_perc['year'] = non_top_1_perc['date'].apply(lambda date: date.year)
non_top_1_perc['month'] = non_top_1_perc['date'].apply(lambda date: date.month)
```

```python
non_top_1_perc = non_top_1_perc.drop('date', axis=1)
```

```python
non_top_1_perc['zipcode'].value_counts()
```

```python
non_top_1_perc = non_top_1_perc.drop('zipcode', axis =1)
```

```python
non_top_1_perc['sqft_basement'].value_counts()
```

```python
X = non_top_1_perc.drop('price', axis =1).values
y = non_top_1_perc['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(1))

model.compile(optimixer='adam', loss='mse')
```

```python
model.fit(x=X_train, y =y_train, validation_data =(X_test, y_test), batch_size=128, epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
mean_absolute_error(y_test, predictions)
```

```python
df['price'].describe()
```

```python
explained_variance_score(y_test, predictions)
```
